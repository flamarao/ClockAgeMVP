# 1. Definição do problema

## 1.1 Descrição do problema

O câncer colorretal é a terceira neoplasia mais incidente e uma das principais causas de mortalidade por câncer no mundo. Embora tradicionalmente considerado uma doença do envelhecimento, nas últimas décadas observa-se uma mudança importante em seu perfil epidemiológico, com aumento consistente da incidência em adultos jovens (<50 anos), enquanto a incidência diminui nas populações mais idosas graças ao rastreamento e à prevenção.

Esse fenômeno sugere que a idade cronológica, isoladamente, não explica a biologia do desenvolvimento tumoral. Evidências recentes indicam que indivíduos mais jovens podem apresentar aceleração do envelhecimento biológico (biological age acceleration), medida por relógios moleculares, associada ao aumento do risco de cânceres de início precoce, especialmente tumores gastrointestinais, incluindo o câncer colorretal.

Esses achados reforçam a hipótese de que diferentes tecidos podem envelhecer em velocidades distintas, refletindo alterações em processos como remodelamento epigenético, dano ao DNA, inflamação crônica, metabolismo e imunossenescência.

Permanece uma importante lacuna científica: embora o envelhecimento biológico esteja associado ao risco de desenvolvimento do câncer, ainda não está claro se, após o estabelecimento do tumor, um maior envelhecimento molecular representa um estado de vulnerabilidade celular decorrente da senescência ou um estado adaptativo caracterizado por inflamação persistente, remodelamento do microambiente tumoral e potencial resistência terapêutica.

Em outras palavras, permanece desconhecido se o envelhecimento, especialemte o tumoral, se constitui de um conjunto de diferentes estados biológicos com implicações distintas para a evolução da doença.

Nesse contexto, a disponibilidade de bases públicas multiômicas, como o The Cancer Genome Atlas (TCGA), aliada ao uso de técnicas de Machine Learning, oferece uma oportunidade para investigar esse paradoxo de forma sistemática. Este trabalho propõe utilizar dados de metilação de DNA e informações clínicas do TCGA-COAD/READ para explorar os determinantes moleculares do envelhecimento tumoral, partindo da hipótese de que o câncer colorretal possa apresentar diferentes estados biológicos de envelhecimento, em vez de um único eixo contínuo de progressão. Embora estudos anteriores tenham utilizado relógios epigenéticos para investigar o envelhecimento biológico em câncer colorretal, principalmente em análises de prognóstico, sobrevida e subtipos moleculares, ainda existem lacunas na compreensão dos determinantes moleculares associados à aceleração da idade epigenética tumoral e de como esses padrões podem ser explorados por modelos interpretáveis de Machine Learning.

O dataset escolhido será composto por dados públicos do The Cancer Genome Atlas (TCGA), especificamente das coortes TCGA-COAD (Colon Adenocarcinoma) e TCGA-READ (Rectum Adenocarcinoma). O TCGA reúne dados clínicos e multiômicos de tumores humanos, incluindo metilação de DNA, expressão gênica, mutações e informações clínicas. Para este MVP, a camada molecular prioritária será a metilação de DNA, por sua relação direta com relógios epigenéticos e envelhecimento biológico.

A fonte principal dos dados será o Genomic Data Commons (GDC) Data Portal e/ou o UCSC Xena Browser, que disponibilizam dados do TCGA em formato acessível para análise computacional. O uso dessas bases é adequado para o MVP porque permite acesso público, reprodutibilidade e integração entre dados clínicos e moleculares.

A escolha de TCGA-COAD/READ se justifica pela relevância clínica do câncer colorretal, pela disponibilidade de dados moleculares públicos e pela relação entre envelhecimento biológico, senescência, inflamação e carcinogênese colorretal. Estudos recentes indicam que o envelhecimento biológico pode estar associado ao risco de câncer gastrointestinal/colorretal, mas ainda há incerteza sobre como interpretar o envelhecimento molecular dentro do tumor: como vulnerabilidade, adaptação, resistência ou combinação desses estados biológicos.

Como os dados do TCGA são públicos, anonimizados e disponibilizados para pesquisa científica, este MVP não envolve identificação individual de pacientes. Ainda assim, as análises serão conduzidas respeitando os termos de uso das bases públicas e reconhecendo as limitações de dados retrospectivos e transversais.

Este trabalho propõe utilizar dados de metilação de DNA e informações clínicas do TCGA-COAD/READ para explorar os determinantes moleculares do envelhecimento tumoral, partindo da hipótese de que o câncer colorretal possa apresentar diferentes estados biológicos de envelhecimento, em vez de um único eixo contínuo de progressão.

Os principais interessados nesta solução são pesquisadores em oncologia translacional, epigenética e biologia do envelhecimento, bioinformatas e cientistas de dados que atuam na integração de dados multiômicos para descoberta de biomarcadores; em perspectiva translacional, oncologistas clínicos, patologistas moleculares e equipes de pesquisa clínica.

A incidência crescente de câncer colorretal em jovens, associada às limitações da idade cronológica como marcador biológico, evidencia a necessidade de novos modelos capazes de compreender a heterogeneidade do envelhecimento tumoral. Compreender esses processos pode contribuir para a identificação de novos biomarcadores relacionados à senescência, inflamação, remodelamento epigenético e adaptação tumoral.

## 1.2 Objetivo do MVP

O objetivo deste MVP é construir e avaliar modelos de Machine Learning para predizer a idade cronológica de pacientes a partir dos perfis de metilação do DNA da coorte TCGA-COADREAD. São comparados um modelo baseline e diferentes modelos supervisionados de regressão, buscando identificar a abordagem com melhor desempenho preditivo e discutir suas limitações e potencial para futuras aplicações em estudos de envelhecimento epigenético tumoral.

Objetivo deste trabalho

O relógio epigenético será utilizado como ferramenta para investigar estados biológicos de envelhecimento tumoral. A predição da idade cronológica representa a etapa inicial do pipeline, enquanto a análise dos resíduos do modelo poderá gerar hipóteses sobre aceleração epigenética e heterogeneidade biológica entre tumores.

## 1.3 Tipo de problema

O problema foi definido como uma tarefa de **regressão supervisionada**, pois o objetivo é prever uma variável numérica contínua — a idade cronológica dos pacientes (`idade_anos`) — utilizando como variáveis preditoras os perfis de metilação do DNA obtidos da plataforma Illumina HumanMethylation450 da coorte TCGA-COADREAD.

## 1.4 Premissas, hipóteses e critérios de sucesso

**Hipóteses iniciais:**
1. Os perfis de metilação do DNA de pacientes com câncer colorretal contêm informação suficiente para predizer parcialmente a idade cronológica dos pacientes por meio de modelos de regressão supervisionada.
2. Os perfis de metilação apresentam padrões associados à idade cronológica, permitindo que algoritmos de Machine Learning identifiquem relações entre os sítios CpG e o envelhecimento.
3. Modelos de regressão com regularização, como Ridge Regression e Elastic Net, apresentarão desempenho superior ao modelo baseline devido à elevada dimensionalidade e à forte correlação entre os sítios CpG.

**Critérios de sucesso:** MAE (Mean Absolute Error) como métrica principal, complementada por RMSE e R². O MVP será considerado bem-sucedido se pelo menos um modelo candidato superar o baseline `DummyRegressor` em MAE, com redução mínima de 10%.

# 5. Preparação dos dados e divisão treino/teste

Nesta etapa é construída a base analítica definitiva a partir dos arquivos clínico e de metilação gerados pelo TCGAbiolinks/R, seguida da divisão treino/teste e da configuração do problema.

```python
# ============================================================
# 5.1.1 - DETERMINACAO DA COORTE ANALITICA
# ============================================================
clinical_path = "/content/drive/MyDrive/TCGA_COADREAD_R/TCGA_COADREAD_clinical.csv"
beta_path = "/content/drive/MyDrive/TCGA_COADREAD_R/TCGA_COADREAD_HM450_beta_values.csv"

clinical = pd.read_csv(clinical_path)
beta_header = pd.read_csv(beta_path, nrows=0)
methylation_samples = list(beta_header.columns[1:])
methylation_patient_ids = set([sample[:12] for sample in methylation_samples])

if "age_at_diagnosis" in clinical.columns:
    clinical["idade_anos"] = (clinical["age_at_diagnosis"] / 365.25).round(1)
    origem_idade = "age_at_diagnosis"
elif "days_to_birth" in clinical.columns:
    clinical["idade_anos"] = (clinical["days_to_birth"].abs() / 365.25).round(1)
    origem_idade = "days_to_birth"
else:
    raise ValueError("Nenhuma variavel de idade encontrada.")

clinical_valid = clinical.dropna(subset=["idade_anos", "bcr_patient_barcode"]).copy()
clinical_patient_ids = set(clinical_valid["bcr_patient_barcode"].astype(str))
common_patient_ids = clinical_patient_ids.intersection(methylation_patient_ids)

resumo_coorte = pd.DataFrame({
    "Etapa": [
        "Registros clinicos totais",
        "Registros clinicos com idade valida",
        "Amostras de metilacao HM450",
        "Pacientes unicos com metilacao",
        "Pacientes em comum clinica + metilacao"
    ],
    "N": [
        clinical.shape[0],
        clinical_valid.shape[0],
        len(methylation_samples),
        len(methylation_patient_ids),
        len(common_patient_ids)
    ]
})
print(f"Origem da idade utilizada: {origem_idade}")
display(resumo_coorte)
```

```python
# ============================================================
# 5.2 - CONFIGURACAO DO PROBLEMA DE MACHINE LEARNING
# ============================================================
TARGET = "idade_anos"
PROBLEM_TYPE = "regressao"

# Colunas que NAO devem entrar como preditoras
ID_COLUMNS = ["patient_id", "sample_id"]
DATE_COLUMN = None
DROP_COLUMNS = []

columns_to_exclude = set(ID_COLUMNS + DROP_COLUMNS)
columns_to_exclude.add(TARGET)

features = [c for c in analysis_df_final.columns if c not in columns_to_exclude]

print("=" * 70)
print("CONFIGURACAO DO PROBLEMA")
print("=" * 70)
print(f"Tipo de problema: {PROBLEM_TYPE}")
print(f"Target: {TARGET}")
print(f"\nNumero de features: {len(features):,}")
print("\nColunas excluidas:", sorted(columns_to_exclude))
```

## 5.3 Construção de X e y

Nesta etapa são construídas a matriz de variáveis preditoras (X) e a variável-alvo (y). **Correção em relação à versão de desenvolvimento:** a conversão de tipos (`.astype`) e qualquer operação `.fillna()`/`.corrwith()` no DataFrame inteiro (485 mil colunas) foram removidas daqui — essas operações fragmentam o gerenciador interno de blocos do pandas e o custo deixa de ser linear, travando a execução. A divisão treino/teste é feita **antes** de qualquer redução de dimensionalidade, e toda a redução (Seção 6) passa a ser feita com ferramentas vetorizadas do scikit-learn, que já haviam se mostrado eficientes (0,58 min) nos testes originais.

```python
# ============================================================
# 5.3 - CONSTRUCAO DE X e y
# ============================================================
X = analysis_df_final[features]
y = analysis_df_final[TARGET]

print("Dimensao de X:", X.shape)
print("Dimensao de y:", y.shape)
```

## 5.4 Justificativa da divisão

Será utilizada a estratégia **holdout**, dividindo a base em treinamento (80%) e teste (20%), resultando em aproximadamente 310 pacientes para treinamento e 78 para teste. Como o problema é de regressão (variável contínua), não é necessária estratificação. Todas as etapas de preparação dependentes da distribuição dos dados (imputação, remoção por variância, seleção de atributos) serão ajustadas **exclusivamente no conjunto de treinamento** e depois aplicadas ao teste, evitando vazamento de dados (*data leakage*).

## 5.5 Divisão treino/teste

```python
# ============================================================
# 5.5 - DIVISAO TREINO/TESTE
# ============================================================
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=SEED
)

print("=" * 70)
print("DIVISAO TREINO/TESTE")
print("=" * 70)
print(f"X_train: {X_train.shape}")
print(f"X_test : {X_test.shape}")
print(f"y_train: {y_train.shape}")
print(f"y_test : {y_test.shape}")
```

```python
# ============================================================
# 5.5.1 - REMOCAO DE CpGs COM ELEVADA PROPORCAO DE VALORES AUSENTES
# (ajustado apenas no conjunto de treinamento)
# ============================================================
missing_threshold = 0.20
missing_rate = X_train.isna().mean()
selected_by_missing = missing_rate[missing_rate <= missing_threshold].index.tolist()

X_train_filt = X_train[selected_by_missing]
X_test_filt = X_test[selected_by_missing]

print("Apos remocao por valores ausentes:")
print("X_train_filt:", X_train_filt.shape)
print("X_test_filt :", X_test_filt.shape)
```

**Nota de correção:** a partir daqui, a imputação de valores ausentes será feita com `SimpleImputer` (Seção 6.1), e não com `.fillna()` do pandas aplicado ao DataFrame inteiro — essa era a segunda operação que travava a execução, pelo mesmo motivo de fragmentação de memória. `SimpleImputer` opera sobre o array NumPy subjacente e resolve a mesma tarefa em segundos.

# 6. Pré-processamento e pipeline

O conjunto de dados contém um número muito elevado de variáveis de metilação em relação ao número de pacientes (485.577 CpGs para 388 pacientes), sendo necessária uma estratégia de redução de dimensionalidade antes do treinamento. O pipeline foi estruturado para evitar vazamento de dados: todas as decisões dependentes da distribuição foram ajustadas exclusivamente no conjunto de treinamento.

Etapas realizadas:
- Remoção de CpGs com elevada proporção de valores ausentes (feita acima, 5.5.1).
- Imputação dos valores ausentes pela mediana do treino.
- Remoção de CpGs com variância muito baixa.
- Seleção supervisionada das 1.000 CpGs mais associadas à idade cronológica.

## 6.1 Pré-processamento eficiente com scikit-learn

```python
# ============================================================
# 6.1 - PRE-PROCESSAMENTO EFICIENTE COM SCIKIT-LEARN
# ============================================================
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import VarianceThreshold, SelectKBest, f_regression
import time

print("=" * 70)
print("PRE-PROCESSAMENTO EFICIENTE")
print("=" * 70)
inicio = time.time()

# 1. Imputacao pela mediana calculada apenas no treino
imputer = SimpleImputer(strategy="median")
X_train_imp = imputer.fit_transform(X_train_filt)
X_test_imp = imputer.transform(X_test_filt)
print("Apos imputacao:", X_train_imp.shape, X_test_imp.shape)

# 2. Remocao de baixa variancia
var_selector = VarianceThreshold(threshold=1e-5)
X_train_var = var_selector.fit_transform(X_train_imp)
X_test_var = var_selector.transform(X_test_imp)
selected_after_variance = X_train_filt.columns[var_selector.get_support()]
print("Apos remocao por baixa variancia:", X_train_var.shape, X_test_var.shape)

# 3. Selecao supervisionada das 1000 CpGs mais associadas a idade
N_FEATURES_FINAL = 1000
feature_selector = SelectKBest(score_func=f_regression, k=N_FEATURES_FINAL)
X_train_selected = feature_selector.fit_transform(X_train_var, y_train)
X_test_selected = feature_selector.transform(X_test_var)
selected_features = selected_after_variance[feature_selector.get_support()].tolist()

fim = time.time()
print("\nApos selecao final de features:")
print("X_train_selected:", X_train_selected.shape)
print("X_test_selected :", X_test_selected.shape)
print("\nPrimeiras 10 CpGs selecionadas:", selected_features[:10])
print(f"\nTempo total: {(fim - inicio)/60:.2f} minutos")
```

**Síntese:** o pipeline reduziu progressivamente a dimensionalidade — de 485.577 CpGs para as CpGs com baixa proporção de ausentes, depois removendo baixa variância, e por fim selecionando as **1.000 CpGs** mais associadas à idade cronológica via `SelectKBest`/`f_regression`. Todas as etapas foram ajustadas apenas no treino e aplicadas ao teste, evitando vazamento de dados.

## 6.2 Decisões de pré-processamento

- **Imputação:** mediana calculada exclusivamente no treino (robusta a outliers, comum em dados biológicos).
- **Remoção por ausência elevada:** evita que CpGs pouco informativas dominem a etapa de imputação.
- **Remoção por baixa variância:** elimina sítios CpG com pouca capacidade discriminativa.
- **Seleção supervisionada (`SelectKBest`/`f_regression`):** com ~485 mil CpGs para 388 pacientes, essa etapa reduz o risco de sobreajuste e o custo computacional.
- **Sem padronização adicional:** os valores beta de metilação já estão na mesma escala (0 a 1).
- **Sem codificação categórica:** todas as variáveis preditoras utilizadas são numéricas.

# 7. Baseline e modelos candidatos

## 7.1 Justificativa dos modelos

Foi utilizado um modelo **baseline** (`DummyRegressor`, prevendo a média), que estabelece o piso de desempenho esperado.

Como modelos supervisionados:
- **Ridge Regression:** regularização L2, adequada para alta dimensionalidade e forte correlação entre CpGs.
- **Elastic Net:** combina regularização L1 (Lasso) e L2 (Ridge), realizando seleção automática de variáveis.
- **Random Forest Regressor:** captura relações não lineares entre metilação e idade cronológica.

O objetivo não foi identificar "o melhor algoritmo" de forma absoluta, mas verificar se diferentes estratégias supervisionadas conseguem extrair, dos perfis de metilação, informação biologicamente relevante para estimar a idade cronológica.

```python
# ============================================================
# 7.2 - MODELO BASELINE
# ============================================================
from sklearn.dummy import DummyRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

baseline = DummyRegressor(strategy="mean")
baseline.fit(X_train_selected, y_train)
baseline_pred = baseline.predict(X_test_selected)

baseline_mae = mean_absolute_error(y_test, baseline_pred)
baseline_rmse = np.sqrt(mean_squared_error(y_test, baseline_pred))
baseline_r2 = r2_score(y_test, baseline_pred)

print(f"MAE : {baseline_mae:.2f} anos")
print(f"RMSE: {baseline_rmse:.2f} anos")
print(f"R2  : {baseline_r2:.3f}")
```

```python
# ============================================================
# 7.3 - RIDGE REGRESSION
# ============================================================
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=1.0)
ridge.fit(X_train_selected, y_train)
ridge_pred = ridge.predict(X_test_selected)

ridge_mae = mean_absolute_error(y_test, ridge_pred)
ridge_rmse = np.sqrt(mean_squared_error(y_test, ridge_pred))
ridge_r2 = r2_score(y_test, ridge_pred)

print(f"MAE : {ridge_mae:.2f} anos")
print(f"RMSE: {ridge_rmse:.2f} anos")
print(f"R2  : {ridge_r2:.3f}")
```

```python
# ============================================================
# 7.4 - RANDOM FOREST
# ============================================================
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=200, random_state=SEED, n_jobs=-1)
rf.fit(X_train_selected, y_train)
rf_pred = rf.predict(X_test_selected)

rf_mae = mean_absolute_error(y_test, rf_pred)
rf_rmse = np.sqrt(mean_squared_error(y_test, rf_pred))
rf_r2 = r2_score(y_test, rf_pred)

print(f"MAE : {rf_mae:.2f} anos")
print(f"RMSE: {rf_rmse:.2f} anos")
print(f"R2  : {rf_r2:.3f}")
```

```python
# ============================================================
# 7.5 - ELASTIC NET
# ============================================================
from sklearn.linear_model import ElasticNet

elastic_net = ElasticNet(alpha=0.01, l1_ratio=0.5, random_state=SEED, max_iter=10000)
elastic_net.fit(X_train_selected, y_train)
elastic_net_pred = elastic_net.predict(X_test_selected)

elastic_net_mae = mean_absolute_error(y_test, elastic_net_pred)
elastic_net_rmse = np.sqrt(mean_squared_error(y_test, elastic_net_pred))
elastic_net_r2 = r2_score(y_test, elastic_net_pred)

print(f"MAE : {elastic_net_mae:.2f} anos")
print(f"RMSE: {elastic_net_rmse:.2f} anos")
print(f"R2  : {elastic_net_r2:.3f}")
```

```python
# ============================================================
# 7.6 - COMPARACAO DOS MODELOS
# ============================================================
comparacao = pd.DataFrame({
    "Modelo": ["Baseline", "Ridge Regression", "Random Forest", "Elastic Net"],
    "MAE": [baseline_mae, ridge_mae, rf_mae, elastic_net_mae],
    "RMSE": [baseline_rmse, ridge_rmse, rf_rmse, elastic_net_rmse],
    "R2": [baseline_r2, ridge_r2, rf_r2, elastic_net_r2]
}).sort_values("RMSE")

display(comparacao)
```

## 7.8 Diagnóstico do modelo candidato inicial (Elastic Net, parâmetros padrão)

Antes de otimizar hiperparâmetros (Seção 9), é útil registrar o comportamento do candidato selecionado **como ele chegou nesta etapa** — com `alpha=0.01` e `l1_ratio=0.5` (valores padrão escolhidos na Seção 7.5, sem busca de hiperparâmetros ainda). Esse diagnóstico documenta o ponto de partida do processo de refinamento e permite comparar, mais adiante, o que a otimização de fato mudou.

```python
# ============================================================
# 7.8 - DIAGNOSTICO DO ELASTIC NET INICIAL (parametros padrao)
# ============================================================
import matplotlib.pyplot as plt

residuals_inicial = y_test - elastic_net_pred

print("Elastic Net inicial (alpha=0.01, l1_ratio=0.5):")
print(f"MAE : {elastic_net_mae:.2f} anos")
print(f"RMSE: {elastic_net_rmse:.2f} anos")
print(f"R2  : {elastic_net_r2:.3f}")

plt.figure(figsize=(6,5))
plt.scatter(elastic_net_pred, residuals_inicial, alpha=0.7)
plt.axhline(0, color="red", linestyle="--")
plt.xlabel("Idade prevista pelo Elastic Net inicial (anos)")
plt.ylabel("Residuo (anos)")
plt.title("Modelo inicial (pre-otimizacao): residuos versus valores previstos")
plt.show()
```

## 7.7 Interpretação dos resultados

# 8. Treinamento e avaliação inicial — síntese

Foram avaliados quatro modelos (baseline, Ridge, Random Forest, Elastic Net) no conjunto de teste, usando MAE, RMSE e R². O Elastic Net apresentou o melhor desempenho geral e foi selecionado para a etapa de otimização de hiperparâmetros. Como esta avaliação já usou o conjunto de teste, novos ajustes daqui em diante são conduzidos por validação cruzada, para evitar viés de seleção.

# 9. Validação e otimização de hiperparâmetros

Foi realizada uma busca em grade (Grid Search) para o Elastic Net, com validação cruzada de 3 folds no conjunto de treinamento, otimizando `alpha` (intensidade da regularização) e `l1_ratio` (equilíbrio entre L1 e L2), guiada pelo MAE negativo.

```python
# ============================================================
# 9 - OTIMIZACAO DE HIPERPARAMETROS DO ELASTIC NET
# ============================================================
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import GridSearchCV

param_grid = {
    "alpha": [0.001, 0.01, 0.1],
    "l1_ratio": [0.2, 0.5, 0.8]
}

elastic_base = ElasticNet(random_state=SEED, max_iter=10000)

grid_elastic = GridSearchCV(
    estimator=elastic_base,
    param_grid=param_grid,
    scoring="neg_mean_absolute_error",
    cv=3,
    n_jobs=-1
)
grid_elastic.fit(X_train_selected, y_train)

best_elastic = grid_elastic.best_estimator_

print("Melhores hiperparametros:", grid_elastic.best_params_)
print(f"Melhor MAE medio na validacao cruzada: {-grid_elastic.best_score_:.2f} anos")
```

**Nota:** o MAE de validação cruzada tende a ser mais conservador do que o MAE observado numa única divisão treino/teste — isso é esperado e reflete uma estimativa mais robusta da capacidade de generalização do modelo.

# 10. Avaliação final no conjunto de teste
Esta seção reavalia o modelo **depois** da otimização de hiperparâmetros (Seção 9), permitindo comparar diretamente com o diagnóstico do candidato inicial apresentado na Seção 7.8.

```python
# ============================================================
# 10.1 - ANALISE DOS RESIDUOS (Elastic Net APOS otimizacao de hiperparametros: best_elastic)
# ============================================================
import matplotlib.pyplot as plt

y_pred_final = best_elastic.predict(X_test_selected)
residuals = y_test - y_pred_final

final_mae = mean_absolute_error(y_test, y_pred_final)
final_rmse = np.sqrt(mean_squared_error(y_test, y_pred_final))
final_r2 = r2_score(y_test, y_pred_final)
print(f"MAE final : {final_mae:.2f} anos")
print(f"RMSE final: {final_rmse:.2f} anos")
print(f"R2 final  : {final_r2:.3f}")

plt.figure(figsize=(6,5))
plt.scatter(y_pred_final, residuals, alpha=0.7)
plt.axhline(0, color="red", linestyle="--")
plt.xlabel("Idade prevista pelo Elastic Net (anos)")
plt.ylabel("Residuo (anos)")
plt.title("Modelo final (pos-otimizacao): residuos versus valores previstos")
plt.show()
```

# 12. Boas práticas e rastreabilidade

| Decisão | Justificativa | Impacto esperado |
|---|---|---|
| `random_state = 42` | Reprodutibilidade | Resultados reproduzíveis |
| Holdout 80/20 | Avaliação independente | Estimativa confiável de generalização |
| Sem estratificação | Alvo contínuo (regressão) | Divisão apropriada |
| Remoção de `patient_id`/`sample_id` | Identificadores não têm valor preditivo | Evita vazamento de informação |
| Conversão da idade de dias para anos | Interpretabilidade | Maior clareza clínica |
| Remoção de CpGs com muitos ausentes | Reduzir ruído | Maior robustez |
| Imputação pela mediana (treino) | Preservar CpGs, evitar leakage | Robustez a outliers |
| Remoção por baixa variância | Eliminar variáveis pouco informativas | Redução de dimensionalidade |
| `SelectKBest`/`f_regression` (1.000 CpGs) | Reduzir dimensionalidade mantendo sinal | Menor custo e menor overfitting |
| **Uso de scikit-learn (não pandas puro) na etapa de imputação/seleção** | **Evitar fragmentação de memória em DataFrames com ~485 mil colunas** | **Pipeline executa em segundos em vez de travar** |
| Grid Search com CV (3 folds) | Escolha objetiva de hiperparâmetros | Maior robustez da seleção |
| MAE, RMSE, R² | Avaliação complementar | Melhor interpretação dos resultados |

**Recursos computacionais:** Google Colab, Python 3, `pandas`/`numpy`/`scikit-learn`/`matplotlib`. Base final: 388 pacientes, 485.577 CpGs inicialmente, 1.000 após seleção.

**Limitações conhecidas:** coorte exclusiva do TCGA-COADREAD; sem validação externa; número de pacientes reduzido frente à dimensionalidade inicial; modelo estima idade cronológica e não constitui, nesta etapa, um relógio epigenético validado.

**Estratégias consideradas e não implementadas:** clusterização hierárquica, ConsensusClusterPlus, PAM e modelos mais complexos (XGBoost, LightGBM, redes neurais) foram deliberadamente adiados por ampliarem o escopo além do necessário para o objetivo deste MVP, permanecendo como perspectivas futuras.
